# **Data Preprocessing**

```python
#converting string values(Categorical Values) to integer
clean_data.Gender=clean_data.Gender.map({"Female":0,"Male":1})
clean_data.Married=clean_data.Married.map({"No":0,"Yes":1})
clean_data.Education=clean_data.Education.map({"Not Graduate":0,"Graduate":1})
clean_data.SelfEmployed=clean_data.SelfEmployed.map({"No":0,"Yes":1})
clean_data.PropertyArea=clean_data.PropertyArea.map({"Rural":0,"Semiurban":1,"Urban":2})
clean_data.LoanStatus=clean_data.LoanStatus.map({"N":0,"Y":1})
```

```python
clean_data.head()
```

```python
# Excluding LoanID
clean_data=clean_data.iloc[:,1:]
clean_data.head()
```

```python
# Seperating the features from the target value
x = clean_data.drop(columns=["LoanStatus"])
y = clean_data["LoanStatus"]
```

```python
# Data Normalization
x = MinMaxScaler().fit_transform(x)
```

```python
# Dividing the data into training set and testing set
# The dataset will be split into 80% train and 20% test.
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=4)
```

# Models

## 1. Logistic Regrression

```python
# Create and train the Logistic Regression model on the training data
model_lr = LogisticRegression()
model_lr.fit(x_train, y_train)

# Make predictions on the test set
y_pred_lr = model_lr.predict(x_test)

# Evaluate the Logistic Regression model
accuracy_lr = accuracy_score(y_test, y_pred_lr)
classification_report_result_lr = classification_report(y_test, y_pred_lr)

print(f"Accuracy: {accuracy_lr}")
print("Classification Report:\n", classification_report_result_lr)

# Cross-validation to evaluate the model's performance on the entire dataset
clf_lr = LogisticRegression()
scores_lr = cross_val_score(clf_lr,x,y,cv=5)
print(f"The accuracy of this model is {scores_lr.mean() :.2%}")
```

```python
# Confusion Matrix
conf_mat_lr = confusion_matrix(y_test, y_pred_lr)
print('Confusion Matrix\n\n',conf_mat_lr)
print('\nTrue Positives(TP) = ', conf_mat_lr[0,0])
print('\nTrue Negatives(TN) = ', conf_mat_lr[1,1])
print('\nFalse Positives(FP) = ', conf_mat_lr[0,1])
print('\nFalse Negatives(FN) = ', conf_mat_lr[1,0])

# Visualization
sns.heatmap(conf_mat_lr,annot= True)
```

## 2. KNN

```python
# Create and train the KNN model on the training data
model_knn = KNeighborsClassifier()
model_knn.fit(x_train, y_train)

# Make predictions on the test set
y_pred_knn = model_knn.predict(x_test)

# Evaluate the KNN model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
classification_report_result_knn = classification_report(y_test, y_pred_knn)

print(f"KNN Accuracy: {accuracy_knn}")
print("Classification Report (KNN):\n", classification_report_result_knn)

# Cross-validation to evaluate the model's performance on the entire dataset
clf_knn = KNeighborsClassifier()
scores_knn = cross_val_score(clf_knn, x, y, cv=5)
print(f"The accuracy of the KNN model is {scores_knn.mean() :.2%}")
```

```python
# Confusion Matrix
conf_mat_knn = confusion_matrix(y_test, y_pred_knn)
print('Confusion Matrix\n\n',conf_mat_knn)
print('\nTrue Positives(TP) = ', conf_mat_knn[0,0])
print('\nTrue Negatives(TN) = ', conf_mat_knn[1,1])
print('\nFalse Positives(FP) = ', conf_mat_knn[0,1])
print('\nFalse Negatives(FN) = ', conf_mat_knn[1,0])

# Visualization
sns.heatmap(conf_mat_knn,annot= True)
```

## 3. Decision Tree

```python
# Create and train the Decision Tree model on the training data
model_dt = DecisionTreeClassifier()
model_dt.fit(x_train, y_train)

# Make predictions on the test set
y_pred_dt = model_dt.predict(x_test)

# Evaluate the Decision Tree model
accuracy_dt = accuracy_score(y_test, y_pred_dt)
classification_report_result_dt = classification_report(y_test, y_pred_dt)

print(f"Decision Tree Accuracy: {accuracy_dt}")
print("Classification Report (Decision Tree):\n", classification_report_result_dt)

# Cross-validation to evaluate the model's performance on the entire dataset
clf_dt = DecisionTreeClassifier()
scores_dt = cross_val_score(clf_dt, x, y, cv=5)
print(f"The accuracy of the Decision Tree model is {scores_dt.mean() :.2%}")
```

```python
# Confusion Matrix
conf_mat_dt = confusion_matrix(y_test, y_pred_dt)
print('Confusion Matrix\n\n',conf_mat_dt)
print('\nTrue Positives(TP) = ', conf_mat_dt[0,0])
print('\nTrue Negatives(TN) = ', conf_mat_dt[1,1])
print('\nFalse Positives(FP) = ', conf_mat_dt[0,1])
print('\nFalse Negatives(FN) = ', conf_mat_dt[1,0])

# Visualization
sns.heatmap(conf_mat_dt,annot= True)
```

## 4. Naive Bayes

```python
# Create and train the Naive Bayes model on the training data
model_nb = GaussianNB()
model_nb.fit(x_train, y_train)

# Make predictions on the test set
y_pred_nb = model_nb.predict(x_test)

# Evaluate the Naive Bayes model
accuracy_nb = accuracy_score(y_test, y_pred_nb)
classification_report_result_nb = classification_report(y_test, y_pred_nb)

print(f"Naive Bayes Accuracy: {accuracy_nb}")
print("Classification Report (Naive Bayes):\n", classification_report_result_nb)

# Cross-validation to evaluate the model's performance on the entire dataset
clf_nb = GaussianNB()
scores_nb = cross_val_score(clf_nb, x, y, cv=5)
print(f"The accuracy of the Naive Bayes model is {scores_nb.mean() :.2%}")
```

```python
# Confusion Matrix
conf_mat_nb = confusion_matrix(y_test, y_pred_nb)
print('Confusion Matrix\n\n',conf_mat_nb)
print('\nTrue Positives(TP) = ', conf_mat_nb[0,0])
print('\nTrue Negatives(TN) = ', conf_mat_nb[1,1])
print('\nFalse Positives(FP) = ', conf_mat_nb[0,1])
print('\nFalse Negatives(FN) = ', conf_mat_nb[1,0])

# Visualization
sns.heatmap(conf_mat_nb,annot= True)
```

# Comparing the models

```python
model_names = ["Logistic Regression", "KNN", "Decision Tree", "Naive Bayes"]
models = [LogisticRegression(), KNeighborsClassifier(), DecisionTreeClassifier(), GaussianNB()]
score = []
for i in model_names :
    clf = models[model_names.index(i)]
    scores = cross_val_score(clf, x, y,cv=5)
    score.append(scores.mean())
    
result = pd.DataFrame({"Model": model_names , "Accuracy":score})
result.sort_values(by="Accuracy", ascending=False)
```

```python
plt.barh(range(len(model_names)), score, align='center', alpha=0.5)
plt.yticks(range(len(model_names)), model_names)
plt.xlabel('Score')
plt.title('Classification Performance')
plt.show()
```

## Hyper parameter tuning

Logistic Regression has the best accuracy among all the models. So, here will be searching best set of hyperparameters for this model for optimal performance.

```python
clf = LogisticRegression(max_iter=10000)

param_grid = {
    'C': np.random.randint(1, 101, size=20)
}

grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, scoring='accuracy', cv=50)

grid_search.fit(x_train, y_train)
```

```python
print(f"Best score: {grid_search.best_score_:.2%}")
print(f"Best parameters: {grid_search.best_params_}")
```
